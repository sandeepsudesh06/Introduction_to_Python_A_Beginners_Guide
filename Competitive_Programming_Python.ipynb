### Python Conditions and statements 

Python supports the usual logical conditions from mathematics:

* Equals: `a == b`
* Not Equals: `a != b`
* Less than: `a < b`
* Less than or equal to: `a <= b`
* Greater than: `a > b`
* Greater than or equal to: `a >= b`

These conditions can be used in several ways, most commonly in "if statements" and loops

## For Loops

### Looping Through an Iterator

Iterate the values in an array

```python
week = ["Sunday", "Monday", "Tuesday"]

for x in week:
    print(x)
```

Iterate the characters in a string

```python
string = "Sunday"

for x in string:
    print(x)
```

```python
# iterating through a range
for x in range(4):
    print(x)
```

```python
for x in range(0,5):
    print(x)
```

## While loop

With the `while` loop we can execute a set of statements as long as a condition is **true**.

```python
x = 1
while x < 6:
    print(x)
    x += 1
```

```python
x = 1
while x < 6:
    print(x)
    if x == 4:
        break
    x += 1
```

## Question 1 - Find even numbers

Find the even number in an array and return a new array with even numbers

```python
# creating a function

def is_even(a):
    
    b=[]
    
    for i in a:
        if i%2==0:  # conditional statement
            b.append(i)
    
    return b
```

```python
a = [1,2,3,4,5,6,7,7]
is_even(a)
```

## Question 2 - Even and Odd

Given an array of positive integers, Print if the number is even or odd

```python
def EvenOdd(a):
    
    for n in a:
        
        if n%2==0:
            print(n , 'is Even')
        else:
            print(n ,'is Odd')
    
```

```python
EvenOdd(a)
```

## Question 3 - Sum of an Array

Given an array of integers, find the sum of its elements.

Example: 

`a = [1, 2, 3]
Output : 6
1 + 2 + 3 = 6`

```python
def total_sum(a):
    
    #write your code here
    sum=0
    for i in a:
        sum = sum + i
          
    return(sum) 
```

```python
a = [10,10,10]
total_sum(a)
```

## Question 4 - Largest element in an array
Given an array, find the largest element in it.

Example:

`a = [10, 20, 30]
Output : 30`

```python
def largest_num(a):
  
    max = a[0]
    for i in range(1, len(a)):
        if a[i] > max:
            max = a[i]
    return max
```

```python
a = [10,20,30]
largest_num(a)
```

## Question 5: Sum of squares of first 'n' natural numbers

Given a positive integer N. The task is to find 1<sup>2</sup> + 2<sup>2</sup> + 3<sup>2</sup> + .. + N<sup>2</sup>

Example:

`Input : N = 4
Output : 30
= 1 + 4 + 9 + 16
= 30`

```python
def square_sum(n) :
    
    s = 0
    for i in range(1, n+1) :
        s = s + (i * i)
      
    return s
```

```python
square_sum(3)
```
